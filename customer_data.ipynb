```python
import pandas as pd
from sqlalchemy import create_engine
df = pd.read_csv(r"C:\Users\Likhitha M\Downloads\customer_shopping_behavior.csv")

# Step 2: PostgreSQL connection details
username = "postgres"
password = "likhitham%4078"   # ✅ '@' replaced with %40
host = "localhost"
port = "5432"
database = "customer_behavior"

# Step 3: Create connection engine
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# Step 4: Table name
table_name = "customer"

# Step 5: Load DataFrame into PostgreSQL
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'")
```
